## Gram-Schmidt Process

```python
v1 = np.array([5,2,4])
v2 = np.array([1,5,6])
v3 = np.array([3,2,5])

#calculations for gram schmidt process
u1 = v1
proj_v2u1 = (np.dot(v2,u1)/ np.dot(u1,u1)) * u1
print("projection of v2 on u1", proj_v2u1)

u2 = v2 - proj_v2u1
proj_v3u1 = (np.dot(v3,u1)/ np.dot(u1,u1)) * u1
proj_v2u2 = (np.dot(v3,u2)/ np.dot(u2,u2)) * u2
u3 = v3 - proj_v3u1 - proj_v2u2
print("u1: ", u1)
print("u2: ", u2)
print("u3: ", u3)
```

```python
fig = plt.figure(figsize=plt.figaspect(0.5))
ax1 = fig.add_subplot(1,2,1, projection='3d')
ax2 = fig.add_subplot(1,2,2, projection='3d')

#starting vectors
ax1.title.set_text('Before Gram Schmidt Process')
ax1.quiver(0, 0, 0, *v1, label='v1', color='r')
ax1.quiver(0, 0, 0, *v2, label='v2', color='g')
ax1.quiver(0, 0, 0, *v3, label='v3', color='b')

#orthogonal vectors
ax2.title.set_text('After Gram Schmidt Process')
ax2.quiver(0, 0, 0, *u1, label='u1', color='r')
ax2.quiver(0, 0, 0, *u2, label='u2', color='g')
ax2.quiver(0, 0, 0, *u3, label='u3', color='b')

ax1.set_xlim([-1.5, 0.5])
ax1.set_ylim([0, 1.5])
ax1.set_zlim([-1, 1])

ax2.set_xlim([-1.5, 0.5])
ax2.set_ylim([0, 1.5])
ax2.set_zlim([-1, 1])


plt.show()
```
