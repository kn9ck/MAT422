# Continunity and Differentiation

## Limits and Continuity
A **limit** describes the behavior of a function as the input approaches a particular value. In calculus, this is essential for defining more advanced concepts like continuity and derivatives. $$\lim_{\textbf{x} \to \textbf{a}}f(\textbf{x})=L $$ A function has a limit
$L$ as $x \to a$ if for any arbitrarily small distance $\epsilon$ from $L$, you can find a small distance $\delta$ such that the function's value is within $\epsilon$ of $L$.

A **continuous** function doesn't have any abrupt changes in value. Mathematically, this means that the limit of the function as $x \to a$ must equal the function's value at $a$. $$\lim_{\textbf{x} \to \textbf{a}}f(\textbf{x})=f(\textbf{a}) $$

```python
import numpy as np
import matplotlib.pyplot as plt

#define function
def f(x):
    return (x**2 - 1) / (x - 1)

# Points near x = 1
x_vals = np.linspace(0.5, 1.5, 100)
y_vals = f(x_vals)

# Plot the function
plt.plot(x_vals, y_vals, label="f(x)")
plt.axvline(x=1, color='r', linestyle='--', label='x = 1')
plt.title('Limit of f(x) as x approaches 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

This function simplifies to $f(x)=x+1$ for all $x \ne 1$, but has a removable discontinuity at $x=1$. The limit as $x \to 1$ exists (which is 2), but the function is not defined directly at $x=1$, showing a difference between **limit** existence and **continuity**.

## Derivatives
The **derivative** measures the rate at which a function changes as its input changes. In the single-variable case, the derivative at $x_0$ is given by the limit: $$f^\prime(x_0)=\lim_{h \to 0} \frac{f(x_0 + h)-f(x_0)}{h}$$ This formula captures how $f$ changes near $x_0$, and it is central to calculus and optimization.

```python
from sympy import symbols, diff
import matplotlib.pyplot as plt
import numpy as np

#define variable and function
x = symbols('x')
f_x = x**2

#derivative
f_prime = diff(f_x, x)
print(f"The derivative of f(x) = x^2 is f'(x) = {f_prime}")

#define function and its derivative
def f(x):
    return x**2

def f_deriv(x):
    return 2*x

x_vals = np.linspace(-3, 3, 100)
y_vals = f(x_vals)

#plot f(x) and tangent line at x = 1
x_tangent = 1
tangent_line = f(x_tangent) + f_deriv(x_tangent) * (x_vals - x_tangent)

plt.plot(x_vals, y_vals, label="f(x) = x^2")
plt.plot(x_vals, tangent_line, label=f"Tangent line at x = {x_tangent}", linestyle='--')
plt.scatter(x_tangent, f(x_tangent), color='red')
plt.title('Derivative and Tangent Line')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

In the example, we calculated the derivative of $f(x)=x^2$, which is $f^\prime(x)=2x$. The derivative gives us the slope of the tangent line at any point on the curve. For example, at $x=1$, the tangent line has a slope of $2$.

## Taylor's Theorem
**Taylor's Theorem** gives a way to approximate a smooth function using a polynomial expansion around a point. Specifically, it approximates $f(x)$ by expressing it as a sum of its derivatives at a point $a$: $$f(b)=f(a)+(b-a)f^\prime(a)+\frac{(b-a)^2}{2!}f^{\prime\prime}(a)+\dots$$ This is a second-order approximation if we stop at the second derivative. The error term measures how closely the polynomial approximates the function beyond this point.

This theorem is a generalization of the **Mean Value Theorem** and provides a useful tool for approximating functions in optimization problems.

```python
import numpy as np
import matplotlib.pyplot as plt

#taylor series approximation for sin(x)
def taylor_sin(x, n_terms):
    sine_approx = 0
    for n in range(n_terms):
        coef = (-1)**n
        sine_approx += coef * x**(2*n + 1) / np.math.factorial(2*n + 1)
    return sine_approx

x_vals = np.linspace(-np.pi, np.pi, 100)
y_true = np.sin(x_vals)
y_approx = taylor_sin(x_vals, 5)

#plot true function and the approximation
plt.plot(x_vals, y_true, label="sin(x)", color="blue")
plt.plot(x_vals, y_approx, label="Taylor approximation (5 terms)", linestyle="--", color="red")
plt.title('Taylor Series Approximation for sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

 We applied Taylor's theorem to approximate $\sin(x)$ using its Taylor series expansion around $x=0$. This example shows how a complicated function like $\sin(x)$ can be approximated with just a few polynomial terms.
